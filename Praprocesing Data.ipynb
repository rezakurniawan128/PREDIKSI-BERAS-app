```python
import pandas as pd
from google.colab import files

# Langkah 1: Upload file
uploaded = files.upload()

# Langkah 2: Ambil nama file yang diunggah
for file_name in uploaded.keys():
    print(f"File yang diunggah: {file_name}")
    file_path = file_name  # Nama file yang diunggah

# Langkah 3: Baca file Excel
df = pd.read_excel(file_path)

# Tampilkan data awal
print("Data Awal:")
print(df.head())

# Hapus kolom yang tidak diperlukan
# Ganti dengan nama kolom yang ingin dihapus
if df.shape[1] > 33:  # Pastikan ada lebih dari 32 kolom
    df = df.iloc[:, :33 ]  # Ambil kolom 0 sampai 31 (kolom ke-32 tidak termasuk)

# Hapus kolom A (kolom pertama) dan baris 1 (baris pertama)
df = df.iloc[0:, 1:]  # Baris mulai dari indeks 1, kolom mulai dari indeks 1

df = df.drop(df.index[0:4])  # Hapus baris 1-4
df = df.drop(df.index[6:])    # Hapus baris 6-terakhir, karena index mulai dari 0

# Simpan data yang sudah dibersihkan ke file baru
cleaned_file_path = "databersihdesember.xlsx"
df.to_excel(cleaned_file_path, index=False)

print("Data yang sudah dibersihkan:")
print(df.head())

# Langkah 5: Unduh file hasil
files.download(cleaned_file_path)
```

DATA TRANSPOSE

```python
import pandas as pd

# Fungsi utama
def main():
    # Meminta pengguna memasukkan nama file
    file_path = input("Masukkan nama file (dengan ekstensi .xlsx): ")

    try:
        # Membaca file Excel
        data = pd.read_excel(file_path, engine='openpyxl')
        print("Dataset Asli:")
        print(data)

        # Melakukan transpose
        data_transposed = data.transpose()

        print("\nDataset Setelah Transpose:")
        print(data_transposed)

        # Simpan hasil ke file baru
        output_file = "data_transposed_desember.xlsx"
        data_transposed.to_excel(output_file, index=False, header=False, engine='openpyxl')
        print(f"\nDataset telah disimpan ke file: {output_file}")

    except FileNotFoundError:
        print("File tidak ditemukan. Pastikan file berada di direktori yang sama dengan skrip.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

if __name__ == "__main__":
    main()
```

CODE TAMBAH KOLOM MONTH DAN TANGGAL

```python
import pandas as pd

# Fungsi utama
def main():
    # Meminta pengguna memasukkan nama file
    file_path = input("Masukkan nama file (dengan ekstensi .xlsx): ")

    try:
        # Membaca file Excel
        data = pd.read_excel(file_path, engine='openpyxl')
        print("Dataset Asli:")
        print(data)

        # Menambahkan kolom "month" dan "Tanggal"
        month_name = "Desember"  # Ganti dengan nama bulan yang sesuai
        data["month"] = month_name
        data["Tanggal"] = range(1, len(data) + 1)

        # Memindahkan kolom "month" dan "Tanggal" ke posisi awal
        columns_order = ["month", "Tanggal"] + [col for col in data.columns if col not in ["month", "Tanggal"]]
        data = data[columns_order]

        print("\nDataset Setelah Menambahkan Kolom:")
        print(data)

        # Simpan hasil ke file baru
        output_file = "data_with_month_and_date_Desember.xlsx"
        data.to_excel(output_file, index=False, engine='openpyxl')
        print(f"\nDataset telah disimpan ke file: {output_file}")

    except FileNotFoundError:
        print("File tidak ditemukan. Pastikan file berada di direktori yang sama dengan skrip.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

if __name__ == "__main__":
    main()
```

CODE GABUNG FILE MENJADI BANYAK SHEET

```python
import pandas as pd
from google.colab import files

# Langkah 1: Upload banyak file Excel
uploaded = files.upload()

# Langkah 2: Ambil daftar file yang diunggah
file_names = list(uploaded.keys())
print("File yang diunggah:", file_names)

# Langkah 3: Tentukan nama file output
output_file = "gabungan_juli-desember.xlsx"

# Langkah 4: Gabungkan file ke dalam satu Excel dengan banyak sheet
with pd.ExcelWriter(output_file) as writer:
    for file_name in file_names:
        df = pd.read_excel(file_name)
        sheet_name = file_name.rsplit('.', 1)[0][:31]  # Maks 31 karakter (batas Excel)
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"File gabungan telah disimpan sebagai {output_file}")

# Langkah 5: Unduh file hasil
display(f"Klik untuk mengunduh: [Download {output_file}](./{output_file})")
```

CODE RESERVE SHEET UNTUK MENGURUTKAN DARI JULI - DESEMBER

```python
import openpyxl

def reverse_sheets_order(file_path, output_path):
    # Load workbook
    wb = openpyxl.load_workbook(file_path)

    # Reverse urutan sheet
    wb._sheets = list(reversed(wb._sheets))

    # Simpan ke file baru
    wb.save(output_path)
    print(f"File berhasil disimpan sebagai {output_path}")

# Contoh penggunaan
file_input = "gabungan_juli-desember.xlsx"  # Ganti dengan nama file yang sesuai
file_output = "hasil_reverse.xlsx"  # File hasil setelah urutan dibalik
reverse_sheets_order(file_input, file_output)
```

CODE MENGGABUNGKAN SHEET MENJADI 1

```python
# Step 1: Upload file
from google.colab import files

# Upload file Excel
uploaded = files.upload()

# Step 2: Import library
import pandas as pd

# Nama file (otomatis sesuai dengan file yang diunggah)
file_name = list(uploaded.keys())[0]

# Step 3: Baca semua sheet
all_sheets = pd.read_excel(file_name, sheet_name=None)  # sheet_name=None membaca semua sheet

# Step 4: Gabungkan semua sheet
combined_data = pd.concat(all_sheets.values(), ignore_index=True)

# Step 5: Simpan hasil ke file baru
output_file = "combined_sheets.xlsx"
combined_data.to_excel(output_file, index=False, sheet_name="Combined_Sheet")

# Step 6: Unduh file hasil
files.download(output_file)

print(f"File hasil '{output_file}' telah selesai dan siap diunduh.")
```
